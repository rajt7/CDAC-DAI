```python
import pandas as pd 
import numpy as np 
from sklearn.decomposition import PCA
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold , GridSearchCV , RandomizedSearchCV

from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('train.csv',index_col=0)
x_test = pd.read_csv('test.csv',index_col=0)
x = df.drop(columns='TARGET')
y = df['TARGET']
```

```python
precom = PCA(n_components=20)
Columns = list(map(lambda x : "PC"+str(x),[i for i in range(1,21)]))
kfold = StratifiedKFold(n_splits=5,shuffle=True, random_state=23)
```

### Random Forest

```python
rf = RandomForestClassifier()

pipe = Pipeline([('PCA',precom),('rf',rf)])

pipe.get_params()

params = {'rf__max_depth':[2,3,4],
          'rf__max_features':[2,3,4],
          'rf__min_samples_split':[2,3,4]}
gcv_rf = GridSearchCV(pipe,param_grid=params,cv = kfold,verbose=2,scoring='neg_log_loss',n_jobs=-4)
gcv_rf.fit(x,y)
```

```python
print(gcv_rf.best_params_)
print(gcv_rf.best_score_)
```

```python
RF_Model = gcv_rf.best_estimator_
y_pred_rf = RF_Model.predict(x_test)
pd.DataFrame({'ID':x_test.index,'TARGET':y_pred_rf}).to_csv('Submission_RF.csv',index=False)
```

* public Score  : 0.5
* private Score : 0.5

### light GBM

```python
LGB_C = LGBMClassifier(random_state=23)

pipe_LGB = Pipeline([('PCA',precom),('LGMB',LGB_C)])

kfold = StratifiedKFold(n_splits=5,shuffle=True, random_state=23)

pipe_LGB.get_params()

params = {'LGMB__max_depth':[2,3,4],
          'LGMB__min_child_samples':[2,3,4],
          'LGMB__learning_rate':[0.1,0.2,0.3]}
gcv_LGBM = GridSearchCV(pipe_LGB,param_grid=params,cv = kfold,verbose=2,scoring='neg_log_loss',n_jobs=-4)
gcv_LGBM.fit(x,y)
```

```python
print(gcv_LGBM.best_params_)
print(gcv_LGBM.best_score_)
```

```python
LGB_Model = gcv_LGBM.best_estimator_
y_pred_LGB = LGB_Model.predict(x_test)
pd.DataFrame({'ID':x_test.index,'TARGET':y_pred_LGB}).to_csv('Submission_LGBM.csv',index=False)
```

* public Score  : 0.49997
* private Score : 0.5

### XG_Boost

```python
XG_C = XGBClassifier(random_state=23)

pipe_XG = Pipeline([('PCA',precom),('XG',XG_C)])

kfold = StratifiedKFold(n_splits=5,shuffle=True, random_state=23)

pipe_XG.get_params()

params = {'XG__n_estimators':[20,30,40],
          'XG__learning_rate':[0.1,0.2,0.3]}
gcv_XG = GridSearchCV(pipe_XG,param_grid=params,cv = kfold,verbose=2,scoring='neg_log_loss',n_jobs=-4)
gcv_XG.fit(x,y)
```

```python
print(gcv_XG.best_params_)
print(gcv_XG.best_score_)
```

```python
XG_Model = gcv_XG.best_estimator_
y_pred_XG = XG_Model.predict(x_test)
pd.DataFrame({'ID':x_test.index,'TARGET':y_pred_XG}).to_csv('Submission_XG.csv',index=False)
```

* public Score  : 0.5
* private Score : 0.5
