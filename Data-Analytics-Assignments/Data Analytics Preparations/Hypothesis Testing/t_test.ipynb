## One Sample t-test
A one-sample t-test is a statistical hypothesis test used to determine if there is a significant difference between the mean of a sample and a known or hypothesized population mean.<br>
Eg:- Consider comparing the heights of students in one class with respect to the whole college.


**Null Hypothesis (H0):** Sample mean (m) is less than or equal to the hypothetical mean. (<=m)<br>
**Alternative Hypothesis (Ha):** Sample mean (m) is greater than the hypothetical mean. (>m)

```python
import numpy as np
from scipy import stats
```

```python
np.random.seed(1)
sample = np.random.normal(150, 10, 20)
sample
```

```python
t_stats, p_value = stats.ttest_1samp(sample, popmean=155)
print(f"T Statistic:", t_stats)
print(f"P value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject the null hypothesis")
    print("There is a significant difference between the sample mean and the " +
          "population mean.")
else:
    print("Do not reject the null hypothesis")
    print("There is no difference between the sample mean and the population mean.")
```

## Two-Sample t-test
A two-sample t-test, also known as an independent-sample test,  compares the means of two independent sample groups. A two-sample t-test aims to compare the means of samples belonging to two different populations. <br>
Eg:- Consider comparing the heights of two classes.

**Null Hypothesis (H0):** Independent sample means (m1 and m2) are equal. (m1=m2)<br>
**Alternative Hypothesis (Ha):** Independent sample means (m1 and m2) are not equal. (m1!=m2)

```python
import numpy as np
from scipy import stats
import pylab
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
sample1 = np.random.normal(30, 16, 50)
sample2 = np.random.normal(33, 18, 50)
```

```python
stats.probplot(sample1, dist="norm", plot=pylab)
plt.show()
```

```python
t_stats, p_value = stats.ttest_ind(a=sample1, b=sample2, equal_var=True)
print(f"T Statistic:", t_stats)
print(f"P Value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject the null hypothesis")
    print("The mean of the first sample is either grreater or less than the mean of the " +
          "second sample.")
else:
    print("Do not reject the null hypothesis")
    print("There is no statistical difference between the mean of sample 1 and sample 2.")
```

## Paired t-test
A paired t-test, also known as a dependent sample test, compares the means of two related samples. The samples belong to the same population and are analyzed under varied conditions, e.g., at different points in time. This test is mostly popular for pretest/posttest type of experiments where a sample is studied before and after its conditions are varied with an experiment. 

**Null Hypothesis (H0):** Dependent sample means (m1 and m2) are equal (m1=m2).<br>
**Alternative Hypothesis (Ha):** Dependent sample means (m1 and m2) are not equal (m1!=m2)

```python
import numpy as np
from scipy import stats
```

```python
np.random.seed(1)
sample1 = np.random.normal(30, 16, 50)
sample2 = np.random.normal(33, 18, 50)
```

```python
t_stats, p_value = stats.ttest_rel(a=sample1, b=sample2)
print(f"T Statistic:", t_stats)
print(f"P Value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject the null hypothesis")
    print("There is a difference between the two mean samples.")
else:
    print("Do not reject the null hypothesis")
    print("The means of two samples from the same population are equal.")
```
